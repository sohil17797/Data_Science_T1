```python
names = ['Sohil', 'Ronak', 'Rahul', 'Smit', 'Pratil']
maths = np.array([90, 88, 90, 85, 70])
eng = np.array([70, 80, 88, 55, 65])

total_marks = maths + eng

plt.bar(names, maths, color='r', label = 'Maths')
plt.bar(names, maths, bottom=eng, color='g', label = 'English')

for i in range(len(names)):
    plt.text(names[i], (eng[i]/2), str(eng[i]), ha='center', color='white', weight='bold')
    plt.text(names[i], (eng[i]+maths[i]/2),str(maths[i]), ha = 'center', color = 'white', weight='bold')
    plt.text(names[i], (total_marks[i]+2), str(total_marks[i]), ha = 'center', color = 'k', weight='bold')
plt.xlabel("Names of Students", color='yellow', fontsize=12,weight='bold')
plt.ylabel("Marks of the Student", color='y', fontsize=12,weight='bold')
plt.title("Total Merks chart of the students", color='b', fontsize=18,weight='heavy')
plt.legend()
plt.tight_layout();
```

```python
name = ['ab', 'cd', 'ef', 'fg', 'gh']
ma = np.array([34,45,54,43,56])
eg = np.array([87,78,67,76,68])

total = ma+eg

plt.bar(name, ma, color='r', label='Maths')
plt.bar(name, ma, bottom=eg, color = 'b', label='English')

for i in range (len(name)):
    plt.text(name[i], eg[i]/2, str(eg[i]), color = 'y', weight='bold')
    plt.text(name[i], eg[i]+ma[i]/2, str(ma[i]), color = 'yellow', weight='bold')
    plt.text(name[i], total[i]+2, str(total[i]), color = 'k', weight='bold')
    
plt.xlabel('Name of the students', color = 'k', fontsize=16, weight='bold')
plt.ylabel('Total Marks', color = 'k', fontsize=16, weight='bold')
plt.legend()
plt.title("Total Marks of the Students", color='g', fontsize=25, weight='bold')

```

```python
country = ['India', 'China', 'USA', 'Russia', 'Japan']
Gold_medals = np.array([12,11,10,9,8]) 
Silver_medals = np.array([14,10,9,8,7])
Bronze_medals = np.array([15,9,8,7,6])

total_medals = Gold_medals + Silver_medals + Bronze_medals

plt.bar(country, Gold_medals, color='#FFD700', label = 'Gold')
plt.bar(country, Silver_medals, bottom=Gold_medals, color='#C0C0C0', label= 'Silver')
plt.bar(country, Bronze_medals, bottom = Gold_medals + Silver_medals, color='#CD7F32', label='Bronze')

for i in range(len(country)):
    plt.text(country[i], Gold_medals[i]/3, str(Gold_medals[i]), ha='center', color = 'k', weight = 'bold')
    plt.text(country[i], Gold_medals[i]+Silver_medals[i]/3, str(Silver_medals[i]), ha='center', color = 'k', weight = 'bold')
    plt.text(country[i], Gold_medals[i]+Silver_medals[i]+Bronze_medals[i]/3, str(Bronze_medals[i]), ha='center', color = 'k', weight = 'bold')
    plt.text(country[i], total_medals[i]+3, str(total_medals[i]), ha='center', color = 'k', weight = 'bold')

plt.xlabel('Country Name',color='b',fontsize=15)
plt.ylabel('Number of Medals',color='b',fontsize=15)
plt.title('Country Wise Medals Report',color='r',fontsize=15)

plt.legend();
```

```python
country_name = ['India','China','US','UK']
gold_medals = np.array([60,40,50,20])
silver_medals = np.array([50,30,25,43])
bronze_medals = np.array([55,24,45,6])
xpos = np.arange(len(country_name)) #[0,1,2,3]
w = 0.2

plt.bar(xpos,gold_medals,color='#FFD700',width=w)
plt.bar(xpos+w,silver_medals,color='#C0C0C0',width=w)
plt.bar(xpos+2*w,bronze_medals,color='#CD7F32',width=w)

plt.xticks(xpos+w,country_name)
plt.ylabel('Country Name',color='b',fontsize=15)
plt.xlabel('Number of Medals',color='b',fontsize=15)
plt.title('Country Wise Medals Report',color='r',fontsize=15)
plt.legend(['gold','silver','bronze'])
for i in range(len(country_name)):
    plt.text(xpos[i],gold_medals[i]+1,gold_medals[i],ha='center',color='r',weight=1000)
    plt.text(xpos[i]+w,silver_medals[i]+1,silver_medals[i],ha='center',color='r',weight=1000)
    plt.text(xpos[i]+2*w,bronze_medals[i]+1,bronze_medals[i],ha='center',color='r',weight=1000)
```
